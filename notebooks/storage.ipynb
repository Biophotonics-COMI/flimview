```python
import flimview.flim as flim
import flimview.io_utils as io
import flimview.models as models
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tnrange, tqdm_notebook, tqdm
from matplotlib import cm
import h5py
import os
import json
```

```python
sdtfile = '/home/matias/Research/GSK/Psoriasis/test_dataset/FLIM_D/60_D_760nm_c11.sdt'
```

```python
data,header = io.read_sdt_file(sdtfile)
```

```python
c11 = flim.FlimCube(data, header)
```

```python
c11.mask_intensity(100)
```

```python
plt.imshow(c11.intensity)
plt.colorbar()
```

```python
c11_bin = flim.binCube(c11, bin=4, kernel='flat')
```

```python
plt.imshow(c11_bin.intensity)
plt.colorbar()
```

```python
io.saveCube(c11)
```

```python
io.saveCube(c11_bin)
```
