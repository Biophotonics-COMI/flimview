```python
from flimview import flim
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
bbin = 4
```

```python
# linear
k1 = flim.getKernel(bin=bbin, kernel='linear')
plt.imshow(k1)
plt.title('linear')
plt.colorbar()
```

```python
# gauss
k2 = flim.getKernel(bin=bbin, kernel='gauss')
plt.imshow(k2)
plt.title('gauss, sigma=(bin+1)/3')
plt.colorbar()
```

```python
# gauss, kernels
k3 = flim.getKernel(bin=bbin, kernel='gauss', sigma=8)
plt.imshow(k3)
plt.title('gauss')
plt.colorbar()
```

```python
# gauss, kernels
k4 = flim.getKernel(bin=bbin, kernel='gauss', sigma=2)
plt.imshow(k4)
plt.title('gauss')
plt.colorbar()
```

```python
# airy disk
k5 = flim.getKernel(bin=bbin, kernel='airy')
plt.imshow(k5)
plt.title('airy')
plt.colorbar()
```

```python
# airy disk
k6 = flim.getKernel(bin=bbin, kernel='airy', sigma=8)
plt.imshow(k6)
plt.title('airy')
plt.colorbar()
```
